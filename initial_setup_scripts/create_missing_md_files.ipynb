```python
import os

# Directory to save markdown files
output_dir = 'missing_md_files'

# Create directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

all_filenames = []

# Function to create markdown file for an entry
def create_markdown_file(entry):
    # Convert entry to lowercase and replace spaces with underscores
    filename = entry.lower().replace(' ', '_') + '.md'
    #filename)

    all_filenames.append(f"- {entry}: {filename}")
    filepath = os.path.join(output_dir, filename)
    
    # File content
    content = f"""---
title: {entry}
---

# {entry}
"""
    
    # Write content to file
    with open(filepath, 'w') as file:
        file.write(content)

# Create a markdown file for each entry
for entry in l:
    create_markdown_file(entry)
```
